```python
# To search a key in a BST
def search(root, key):
    
    if root is None or root.val == key:
        return root
    
    if key > root.val:
        return search(root.right, key)
    
    # if key < root.val
    return search(root.left, key)
```

```python
class Node:
    def __init__(self, key):
        # node values, in left and right side
        self.left = None
        self.right = None
        # node value
        self.val = key
        
def insert(root, key):
    if root is None:
        return Node(key)
    else:
        # Esse pedaco justament eh utilizado nas condicoes que seguem ao if
        # para efetivamente mudar o valor
        # que justamente chegamos ate o ponto em que o a chave é igual ao valor do nó raiz
        # passado para a função insert
        if root.val == key:
            return root
        elif key > root.val:
                # if key value is greater than root value, 
                # insert the key at right position
                root.right = insert(root.right, key)
        else:
            root.left = insert(root.left, key)
    return root
        
        
```

## Preorder, Inorder & PostOrder traversal

### Inorder

```python
def inorder(root):
    if root:
        # printa valor do left
        inorder(root.left)
        print(root.val)
        # printa valor do right 
        inorder(root.right)
```
