```python
def display_hash(hashtable):
    
    for i,v in enumerate(hashtable):
        print(f'{i} : {v} \n')
```

```python
# Creating HashTable as nested list
hashtable = [[] for _ in range(10)]
```

```python
hashtable
```

```python
# Hashing function to return 
# key for every value:
# hash function based on length
def hashing(keyvalue):
    key_value = keyvalue
    length = len(hashtable)
    print(key_value)
    print(length)
    return key_value % length
```

```python
# Insert Function do add value to the hash table
def insert(hashtable, key, value):
    
    # hash according to key
    hash_key = hashing(key)
    # return the value generated by hashing key
    # Se cair na mesma hash_key, ele appenda, 
    # a hash_key acaba sendo o index
    hashtable[hash_key].append(value)
```

```python
insert(hashtable, 10, 'Allahabad')
```
