## Function

```python
def hash(self, key:str) -> int:
    hashsum = 0
    
    # for each character in the key
    for idx, c in enumerate(key):
        
        hashsum += (idx + len(key))**ord(c)
        
        # perform modulus do keep hashsum in range [0, self.capacity -1]
        hashsum = hashsum % self.capacity
    return hashsum
```

## Insert

```python
def insert(self, key:str, value) -> None:
    
    # 1. increment size
    self.size +1
    
    # 2. compute index (hash key)
    # the index is exactly the hash_key generated by the hash function
    index = self.hash(key)
    
    node = self.buckets[index]
    
    # if bucket is empty:
    # a node is a bucket
    if node is None:
        # create a node, add it, return 
        
        self.buckets[index] = Node(key,value)
    
    # 4. Collision! Iterate to the end of the linked list at provided index
    
    prev = node
    while node is not None:
        prev = node
        node = node.next
    # Add a new node at the end of the list with provided key/value
    prev.next = Node(key, value)
```

## Find

```python
def find(self, key:str) -> str:
    #1. Compute hash
    index = self.hash(key)
    
    # go to first node in list at bucket
    node = self.buckets[index]
    
    # Traverse the linked list at this node
    while node is not None and node.key != key:
        node = node.next
        
    # 4. Now, node is the nrequested key/value pair or None
    
    if node is None:
        return None
    else:
        return node.value
```

## Remove

```python
def remove(self, key:str) -> str:
    # 1. compute hash 
    
    index = self.hash(key)
    node = self.buckets[index]
    prev = None
    
    # enquanto nao achar meu node
    while node is not None and node.key !=key:
        prev = node
        node = node.next
        
    # Now, node is either the requested node or none
    
    if node is None:
        return None
    else:
        # key was found
        self.size +-1
        # show value
        result = node.value
        
        # Delete this element in linked list
        if prev is None:
            node = None
        else:
            prev.next = prev.next.next
            
        return result
```

# Whole classes

```python
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None
    def __str__(self):
        return "<Node: (%s, %s), next: %s>" % (self.key, self.value, self.next != None)
    def __repr__(self):
        return str(self)
```

```python
# Capacity for internal array
class HashTable:
    def __init__(self, capacity):
        self.capacity = capacity
        self.size = 0
        self.buckets = [None] * self.capacity
    
    def hashkey(self, key:str) -> int:
        hashsum = 0

        # for each character in the key
        for idx, c in enumerate(key):

            hashsum += (idx + len(key))**ord(c)

            # perform modulus do keep hashsum in range [0, self.capacity -1]
            hashsum = hashsum % self.capacity
        return hashsum

    def insert(self, key:str, value) -> None:
    
        # 1. increment size
        self.size +1

        # 2. compute index (hash key)
        # the index is exactly the hash_key generated by the hash function
        index = self.hashkey(key)

        node = self.buckets[index]

        # if bucket is empty:
        # a node is a bucket
        if node is None:
            # create a node, add it, return 

            self.buckets[index] = Node(key,value)
            return

        # 4. Collision! Iterate to the end of the linked list at provided index

        prev = node
        while node is not None:
            prev = node
            node = node.next
            # Add a new node at the end of the list with provided key/value
        prev.next = Node(key, value)
    
    def find(self, key:str) -> str:
        #1. Compute hash
        index = self.hashkey(key)

        # go to first node in list at bucket
        node = self.buckets[index]

        # Traverse the linked list at this node
        while node is not None and node.key != key:
            node = node.next

        # 4. Now, node is the nrequested key/value pair or None

        if node is None:
            print('not found')
        else:
            return node.value
        
    def remove(self, key:str) -> None:
        # 1. compute hash 

        index = self.hashkey(key)
        node = self.buckets[index]
        prev = None

        # enquanto nao achar meu node
        while node is not None and node.key !=key:
            prev = node
            node = node.next

        # Now, node is either the requested node or none

        if node is None:
            print('404')
        else:
            # key was found
            self.size +-1
            # show value
            result = node.value

            # Delete this element in linked list
            if prev is None:
                self.buckets[index] = node.next
            else:
                prev.next = prev.next.next

            return result
```
